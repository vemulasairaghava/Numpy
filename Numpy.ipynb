Indexing and Slicing:
You are working as a Data Analyst at a fruit distribution company.

The company stores the number of fruit boxes shipped per day (for 1 week) in a 2D NumPy array where:
Each row represents a type of fruit.
Each column represents a day of the week (Mon–Sun).


Your Tasks
Create the 2D NumPy array

Create an array named shipments using the above data.
Access single elements (Indexing):
Get the number of Apples shipped on Wednesday.
Get the number of Oranges shipped on Sunday.
Access multiple elements (Slicing):
Extract all the weekday shipments (Mon–Fri) for all fruits.
Extract only the weekend shipments (Sat & Sun).
Extract the shipments of Apples from Wednesday to Saturday.
Bonus (Practice thinking):
Replace the Saturday shipment of Apples (currently 80) with 100 using indexing.


```python
shipments=np.array([[50,60,55,70,65,80,90],[40,45,50,60,55,70,75]])
print(shipments[0,2:3])
```

```python
#  Oranges shipped on Sunday
print(shipments[1,6])
```

```python
print(shipments[:,0:5])
```

```python
print("Extract only the weekend shipments (Sat & Sun)")
print(shipments[:,5:])
```

```python
print("Extract the shipments of Apples from Wednesday to Saturday.")
print(shipments[0,2:-1])
```

```python
print("Replace the Saturday shipment of Apples (currently 80) with 100 using indexing.")
shipments[0:5,6]=100
shipments[1:5,6]=75
print(shipments)
```

```python
farmA=np.array([[10,12,13],[9,11,10],[12,13,14],[8,7,9]])
farmB=np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC=np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])
production_all=np.stack([farmA,farmB,farmC])
print(production_all)
```

```python
print(production_all.ndim)
print(production_all.shape)
print(production_all.size)
print(production_all.dtype)
```

```python
total_per_farm = np.sum(production_all, axis=(1, 2))

mean_per_farm = np.mean(production_all, axis=(1, 2))
std_per_farm = np.std(production_all, axis=(1, 2))
var_per_farm = np.var(production_all, axis=(1, 2))

print("Total per farm:", total_per_farm)
print("Mean per farm:", mean_per_farm)
print("Std per farm:", std_per_farm)
print("Var per farm:", var_per_farm)
```

```python


# Per-hour matrix multiplication idea (matrix op)
# For a quick weighted estimate, create a hour_weights row vector of shape (3,) = [0.5, 1.0, 0.75] (weights for hour1, hour2, hour3).
# Compute weighted_per_panel by performing a dot product between each farm's panel-row and hour_weights to get per-panel weighted energy. (Use np.dot / matrix multiplication appropriately.)

hour_weights = np.array([0.5, 1.0, 0.75])

weighted_per_panel = np.array([np.dot(farm, hour_weights) for farm in production_all])
print("Weighted per panel (each farm):\n", weighted_per_panel)
```

```python
production_all[1] = production_all[1].astype(float) + 0.5
print("After calibration offset (Farm B):\n", production_all[1])
```

```python
# reshape means and stds for broadcasting
means = mean_per_farm.reshape(3, 1, 1)
stds = std_per_farm.reshape(3, 1, 1)

normalized = (production_all - means) / stds
print("Normalized data (first farm):\n", normalized[0])
```

```python
exp_data = np.exp(normalized)
expm1_data = np.expm1(normalized)

# To avoid log on negative/zero values, shift
log_data = np.log(np.abs(normalized) + 1)
log10_data = np.log10(np.abs(normalized) + 1)
log2_data = np.log2(np.abs(normalized) + 1)

sin_data = np.sin(normalized)
cos_data = np.cos(normalized)
tan_data = np.tan(normalized)

arcsin_data = np.arcsin(np.clip(np.tanh(normalized), -1, 1))  # safe mapping
arccos_data = np.arccos(np.clip(np.tanh(normalized), -1, 1))
arctan_data = np.arctan(normalized)

deg_data = np.degrees(normalized)
rad_data = np.radians(normalized)

round_data = np.round(normalized)
floor_data = np.floor(normalized)
ceil_data = np.ceil(normalized)
trunc_data = np.trunc(normalized)
```

```python
high_output = production_all > 20

# Example extra mask: even hour index (hours 0,2)
even_hours_mask = np.arange(3) % 2 == 0
even_hours_mask = even_hours_mask.reshape(1, 1, 3)

high_and_even = np.logical_and(high_output, even_hours_mask)
high_or_even = np.logical_or(high_output, even_hours_mask)
not_high = np.logical_not(high_output)
xor_mask = np.logical_xor(high_output, even_hours_mask)

ge_mask = np.greater_equal(production_all, 20)
less_mask = np.less(production_all, 10)
```
